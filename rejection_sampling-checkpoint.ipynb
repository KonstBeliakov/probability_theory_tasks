Let's make a uniform sampler from the unit disc centered at the origin using rejection sampling. 2 base points plus at most 3 bonus points.

```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng() # set up the random number generator
```

We first write a simple rejector. In this basic approach we simply throw away all points outside the unit disc and do _not_ resample them until we have the required number of points sampled.

```python
def rejection_from_sample(sample: np.ndarray) -> np.ndarray:
    '''
    Given a sample of points in some domain, return only those points which lie within the unit disc centered at the origin.
    :param sample:
    :return:
    '''
    return sample[np.linalg.norm(sample, axis=1) <= 1]
```

Now let's write two samplers: one uniform in the square [-1, 1] x [-1, 1], and another uniform in the regular octagon circumscribed around the unit disc. They should return a `(num_samples, 2)` array of floats.

The rules:
1. Pure Python with numpy, no other libraries.
2. Use the provided `rng` random number generator.
3. Use only the uniform samplers from `rng`: both continuous (`rng.uniform`) and discrete (`rng.integers`).

```python
from random import uniform

def square_sample(num_samples: int) -> np.ndarray:
    """
    Generate points uniformly from the square [-1, 1] x [-1, 1].
    """
    return np.array([(uniform(-1, 1), uniform(-1, 1)) for _ in range(num_samples)])

def point_in_octagon(x: float, y: float) -> bool:
    return abs(x) + abs(y) <= np.sqrt(2)

def octagon_sample(num_samples: int) -> np.ndarray:
    """
    Generate points uniformly from the regular octagon circumscribed around the unit disc.
    :param num_samples:
    :return: np.ndarray of coordinates of points
    """
    points = np.zeros((num_samples, 2))
    c = 0
    while c < num_samples:
        x = uniform(-1, 1)
        y = uniform(-1, 1)
        if point_in_octagon(x, y):
            points[c][0] = x
            points[c][1] = y
            c += 1
    return np.array(points)
```

pNow let's check visually that samplers work as intended and record the acceptance rate.

```python
size = 10**4

square_pts = square_sample(size)
disc_pts = rejection_from_sample(square_pts)

fig, axes = plt.subplots(1, 3, figsize=(14, 4))
ax0, ax1, ax2 = axes
ax0.scatter(square_pts[:,0], square_pts[:,1], marker='.', s=1)
ax1.scatter(disc_pts[:,0], disc_pts[:,1], marker='.', s=1)
ax2.hist2d(disc_pts[:,0], disc_pts[:,1], bins=50, density=True)
plt.show()

print(f'Acceptance rate for square sampler: {disc_pts.shape[0] / size:.3f}')
```

```python
size = 10**4

octagon_pts = octagon_sample(size)
disc_pts = rejection_from_sample(octagon_pts)

fig, axes = plt.subplots(1, 3, figsize=(14, 4))
ax0, ax1, ax2 = axes
ax0.scatter(octagon_pts[:,0], octagon_pts[:,1], marker='.', s=1)
ax1.scatter(disc_pts[:,0], disc_pts[:,1], marker='.', s=1)
ax2.hist2d(disc_pts[:,0], disc_pts[:,1], bins=50, density=True)
plt.show()

print(f'Acceptance rate for octagon sampler: {disc_pts.shape[0] / size:.3f}')
```
